# How to use PCAtomography package

## Basic functions and how to run the PCA tomography

```python
#importing
from PCAtomography import IFUcube
import os #optional
```

```python
#loading a IFU data cube (.fits file)
folder = './cubes/'
cube_name = 'ngc3081_HYPERCUBE.fits' 
fname = os.path.join(cwd, cube_name)

#Instantiating IFUcube object
cube = IFUcube(fname)
```

### Running PCA tomography

```python
cube.run_pca(write_fits = True, out_ext='_PCA')
```

You can inspect the components tomograms and Eigenvectors by using '.plot_tomogram(component)' method,
the component numbers stars at 1.

```python
cube.plot_tomogram(1)
```

```python
cube.plot_tomogram(10)
```

The scree test plot (see Steiner et al. 2009, for details) can be plot with '.plot_scree_test()'

```python
cube.plot_scree_test(components = 20)
```

When usefull you can load a fits resulting from the '.run_pca()', by using '.read_pca()'

```python
PCA_cube_name = 'ngc3081_HYPERCUBE_PCA.fits'

cube = IFUcube.read_pca(os.path.join(cwd, PCA_cube_name))
```

You can get a component tomogram or a eigenvector by doing:

```python
tomogram_1 = cube.get_tomogram(1)

eigenvector_20 = cube.get_eigenvector(20)
print(eigenvector_20)
```

# 'Instrumental Fingerprints' Removal

The .clean_data() method allows the removal of components from the data cube, it is particularly useful for 'instrumental fingerprint' removal from IFU cubes, the methods is described in Menezes et al. 2018 section 6: https://academic.oup.com/mnras/article/438/3/2597/973582#91824382b

When one is interested in removing instrumentral effects on the sensitivity of the CCD that were not removed in the reduction process it is instersting to apply the PCA tomography with the emission lines masked (or interpolated).

```python
#Therefore one need to create a emission line mask first (0 or False values for the emission lines)

#accesing the cube wavelenghts
wl = cube.wl

mask = (wl > 5740 & wl < 5950) or \
    (wl > 5910 & wl < 5950) or \
    (wl > 6110 & wl < 6155) or \
    (wl > 6540 & wl < 6380) or \
    (wl > 6400 & wl < 6440) or \
    (wl > 5580 & wl < 6650) or \
    (wl > 6750 & wl < 6800)

mask = ~mask

cube.run_pca(emission_line_mask=mask, emission_line_treatment='interpolate')
```
